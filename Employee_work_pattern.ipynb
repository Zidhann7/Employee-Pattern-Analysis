## Import csv Data Set

```python
import pandas as pd
df = pd.read_csv('emp.csv')
df
```

## Identify null values in each column

```python
df.isna().sum()
```

## Replace Missing Age with average age

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

## Replace Missing Experience Years With Median

```python
df['Experience_Years'].fillna(df['Experience_Years'].median(),inplace=True)
```

## Fill Missing Job Satisfaction With Mode

```python
df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mode(),inplace=True)
```

## Remove Duplicate Values

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns
```


## Replace Missing Avg Daily Hours with Average 

```python
df['Avg_Daily_Hours'].fillna(df['Avg_Daily_Hours'].mean(),inplace=True)
```

```python
df
```

## Replace Missing Value Project Completed with Average 

```python
df['Projects_Completed'].fillna(df['Projects_Completed'].mean(),inplace=True)
```

## Replace Missing Performance Sore with Average

```python
df['Performance_Score'].fillna(df['Performance_Score'].mean(),inplace=True)
```

## Replace Missing Sick Days With 0

```python
df['Sick_Days'].fillna(0,inplace=True)
```

## Replace Missing Training Hours With Average

```python
df['Training_Hours'].fillna(df['Training_Hours'].mean(),inplace=True)
```

```python
df.isna().sum()
```

## Convert Avg Daily Hours to Int

```python
df['Avg_Daily_Hours'] = df['Avg_Daily_Hours'].astype(int)
```

```python
df['Productivity_Index'] = df['Projects_Completed']*df['Performance_Score']/100
```

```python
df
```

## Create New Column Performance Level 

```python
def pl(score):
    if score < 75:
        return 'Low'
    elif score >= 75 and score <= 85:
        return 'Medium'
    else:
        return 'High'

df['Performance_Level'] = df['Performance_Score'].apply(pl)
```
